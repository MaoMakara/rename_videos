```python
import os
import cv2
import numpy as np

# Set the directory path to your video files
input_dir = r'C:\Users\Mao MAKARA\source\repos\CPD subject\train\Silk'
output_dir = r'C:\Users\Mao MAKARA\source\repos\CPD subject\train\VSilk'

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all files in the directory
for filename in os.listdir(input_dir):
    if filename.endswith(".mp4"):
        # Read the video file
        video = cv2.VideoCapture(os.path.join(input_dir, filename))
        
        # Get the video width and height
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # Get the video fps and number of frames
        fps = int(video.get(cv2.CAP_PROP_FPS))
        num_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        
        # Generate a random angle between 20 and 50 degrees
        angle = 50
        rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
        
        # Create a VideoWriter object to save the rotated video
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        output_filename = os.path.join(output_dir, filename)
        out = cv2.VideoWriter(output_filename, fourcc, fps, (width, height))
        
        # Loop through all frames in the video
        for i in range(num_frames):
            ret, frame = video.read()
            if not ret:
                break
            
            # Apply the rotation to the frame
            rotated_frame = cv2.warpAffine(frame, rotation_matrix, (width, height))
            
            # Write the rotated frame to the output file
            out.write(rotated_frame)
            
        # Release the VideoCapture and VideoWriter objects
        video.release()
        out.release()
```
